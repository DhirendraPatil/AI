```python
from datasets import load_dataset
#load the indian cuisine mini data from the hugginface dataset
dataset = load_dataset("VishalMysore/newIndianCuisine", split="train")
```

```python
dataset[0]
```

```python
import pandas as pd
df = pd.DataFrame(dataset)
```

```python
#converting the pre-processed dataframe back to dataset
dataset = dataset.from_pandas(df)
```

```python
# Now import chromadb, instantiate a chroma client and create a collection using the function create_collection, give it a name.
# you can specify the embedding function to be used as a parameter in the create_collection(), if you don't, a default technique is used.
import chromadb
client = chromadb.Client()
collection = client.create_collection("cookGPTMini1")
```

```python
# Now we will store the Ingredients field of first 100 recipe from dataset into collection.
collection.add(
    ids=[str(i) for i in range(0, 100)],  # every entry in collection requires an ID, ID is a string
    documents=dataset["Ingredients"][:100],# Ingredients field text, for first 100 questions only
    metadatas=[{"type": "Ingredients"} for _ in range(0, 100)],) #metadata, additional information about the documents
```

```python
collection.peek() # peek() returns a list of the first 10 items in the collection
```

```python
results = collection.query(query_texts=dataset["Recipe"][:10], n_results=1)
print(results)
```

```python
results = collection.query(query_texts="Shrimp", n_results=1)
print(results)
```

```python
results = collection.query(query_texts="Paneer", n_results=1)
print(results)
```

```python
results = collection.query(query_texts="Alien", n_results=1)
print(results)
```

```python
# looking at first 5 recipe corresponding to the results returned by collection.query()
dataset['Recipe'][:5]
```

```python
# Now, we will print the first 5 recioes and their ingredients as returned by the query()
for i, q in enumerate(dataset['Recipe'][:5]):
    print(f"Recipe: {q}")
    print(f"Ingredients: {results['documents'][i][0]}")
    print()
```
